```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("pd_data_initial_preprocessing.csv")
df.head(10)
```

```python
df.dtypes
```

```python
variables_continuas = df.columns
matriz_correlacion = df[variables_continuas].corr()
sns.heatmap(matriz_correlacion, annot=False, center=0)
plt.show()
```

```python
# Contar el nÃºmero total de valores en cada columna
total_values = df.shape[0]

# Calcular el nÃºmero de valores missing en cada columna
num_missing = df.isnull().sum()

# Calcular el porcentaje de valores missing en cada columna
percent_missing = num_missing / total_values * 100
percent_missing

```

```python
cat_features = (df.dtypes[df.dtypes == 'object'].index.to_list())
num_features = (df.dtypes[df.dtypes != 'object'].index.to_list())
df[cat_features] = df[cat_features].fillna(df.mode().iloc[0])
df[num_features] = df[num_features].fillna(-99)
```

Hemos rellenado las variables discretas con la moda de cada una ya que estas variables no tenian mas de un 5% de los datos NAs. Para las variables continuas hemos rellenado los NAs con el valor -99 que es muy distinto al resto ya quetienen un porcentaje muy alto

```python
df.head()
```

```python
#Vemos que esta capada por lo tanto quitamos esa variable del dataframe ya que no nos da ninguna informacion
df['zip_code'].head() 
```

```python
df= df.drop(['url', 'zip_code'], axis=1)
```

```python
df = df.set_index(['id'])
df
```

```python
df['loan_status'].value_counts()
```

```python
# Agrupar los valores "Charged off" y "Default" en una nueva categorÃ­a "Not Paid"
df['loan_status'] = df['loan_status'].replace(['Charged Off', 'Default'], 'Not Paid')

# Filtrar los valores "Fully Paid" y "Not Paid" en un nuevo DataFrame
data = df.loc[df['loan_status'].isin(['Fully Paid', 'Not Paid'])]
```

```python
data.shape
```

```python
print(data['emp_title'].unique().shape[0])
data['emp_title'].value_counts()
```

```python
data = data.drop('emp_title', axis=1)
```

Podemos ver la variable emp_title, que te dice The job title supplied by the Borrower when applying for the loan, tiene 125,000 valores distintos. luego vemos qe hay muchos valores que aprecen solo una vez mientras que hay otros como teacher que aparecen mas de 30,000 veces. tendremos qu valorar si la eliminamos o no y finalmente lo hacemos ya que no sabemos como normalizar sino

```python
#vemos que esta muy desbalanceado y por tanto tendremos que balancearlo
data['loan_status'].value_counts(normalize = True)
```

```python
#cogemos 10,000 de Fully paid y 10,000 de Not Paid para balancear la variable objetivo
positive_df = data[data['loan_status'] == 'Fully Paid'].sample(n=10000, random_state=42, replace=True)
negative_df = data[data['loan_status'] == 'Not Paid'].sample(n=10000, random_state=42, replace=True)
#concatenamos los dos dataframes
loan_df = pd.concat([positive_df, negative_df], axis = 0)
#lo barajamos
loan_df = loan_df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
from sklearn.preprocessing import LabelEncoder

# Transformamos la variable objetivo cambiando Fully paid por un 0 y Not Paid por un 1
le = LabelEncoder()
loan_df['loan_status'] = le.fit_transform(loan_df['loan_status'])

# Transformar las variables categÃ³ricas a numericas
cat_vars = ['grade', 'sub_grade', 'home_ownership', 'verification_status', 'purpose', 'addr_state', 'initial_list_status', 'application_type']
for var in cat_vars:
    le = LabelEncoder()
    loan_df[var] = le.fit_transform(loan_df[var])

# Ver los datos transformados
loan_df.head()
```

```python
print(loan_df.shape)
loan_df['loan_status'].value_counts(normalize = True)
```

Vemos que se ha quedado un dataset con 20,000 columnas y 47 variables y la variable loan_status esta balanceada 50-50

```python
from sklearn.model_selection import train_test_split

# Separar la variable objetivo
y = loan_df['loan_status']

# Separar las variables predictoras
X = loan_df.drop('loan_status', axis=1)

# Separar en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## RegresiÃ³n LogÃ­stica

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# Crear objetos para escalar los datos
scaler = StandardScaler()

# Ajustar el escalador en el conjunto de datos de entrenamiento
scaler.fit(X_train)

# Normalizar los datos de entrenamiento y test
X_train_norm = scaler.transform(X_train)
X_test_norm = scaler.transform(X_test)
```

```python
# Entrenamos el modelo de RegresiÃ³n LogÃ­stica con los datos de entrenamiento
logreg = LogisticRegression()
logreg.fit(X_train_norm, y_train)
```

```python
# Importar las mÃ©tricas
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, f1_score, roc_curve, auc, balanced_accuracy_score

# Evaluar precisiÃ³n del modelo en el conjunto de entrenamiento
train_score = logreg.score(X_train_norm, y_train)
print('PrecisiÃ³n en conjunto de entrenamiento:', train_score)

# Evaluar precisiÃ³n del modelo en el conjunto de prueba
test_score = logreg.score(X_test_norm, y_test)
print('PrecisiÃ³n en conjunto de prueba:', test_score)

# Predecir las clases del conjunto de prueba
y_pred = logreg.predict(X_test_norm)

# Calcular mÃ©tricas de clasificaciÃ³n
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
balanced_accuracy = balanced_accuracy_score(y_test, y_pred)
confusion = confusion_matrix(y_test, y_pred)

# Calcular curva ROC y Ã¡rea bajo la curva
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)

# Imprimir resultados
print('\nMÃ©tricas de evaluaciÃ³n:')
print('Accuracy:', accuracy)
print('Precision:', precision)
print('Recall:', recall)
print('Confusion Matrix:\n', confusion)
print('F1-Score:', f1)
print('ROC AUC:', roc_auc)
print('Balanced Accuracy:', balanced_accuracy)
```

```python
data1 = pd.read_csv("pd_data_initial_preprocessing.csv")
```

```python
# Agrupar los valores "Charged off" y "Default" en una nueva categorÃ­a "Not Paid"
data1['loan_status'] = data1['loan_status'].replace(['Charged Off', 'Default'], 'Not Paid')

# Filtrar los valores "Fully Paid" y "Not Paid" en un nuevo DataFrame
data1 = data1.loc[data1['loan_status'].isin(['Fully Paid', 'Not Paid'])]
```

```python
#cogemos 10,000 de Fully paid y 10,000 de Not Paid para balancear la variable objetivo
positive_df1 = data1[data1['loan_status'] == 'Fully Paid'].sample(n=10000, random_state=42, replace=True)
negative_df1 = data1[data1['loan_status'] == 'Not Paid'].sample(n=10000, random_state=42, replace=True)
#concatenamos los dos dataframes
loan_df1 = pd.concat([positive_df1, negative_df1], axis = 0)
#lo barajamos
loan_df1 = loan_df1.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
loan_df1= loan_df1.drop(['url', 'zip_code', 'emp_title'], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
# Transformamos la variable objetivo cambiando Fully paid por un 0 y Not Paid por un 1
le = LabelEncoder()
loan_df1['loan_status'] = le.fit_transform(loan_df1['loan_status'])

# Transformar las variables categÃ³ricas a numericas
cat_vars = ['grade', 'sub_grade', 'home_ownership', 'verification_status', 'purpose', 'addr_state', 'initial_list_status', 'application_type']
for var in cat_vars:
    le = LabelEncoder()
    loan_df1[var] = le.fit_transform(loan_df1[var])
```

```python
from sklearn.model_selection import train_test_split

# Separar la variable objetivo
y1 = loan_df1['loan_status']

# Separar las variables predictoras
X1 = loan_df1.drop('loan_status', axis=1)

# Separar en conjunto de entrenamiento y prueba
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.3, random_state=42)
```

```python
from sklearn.impute import KNNImputer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.2, random_state=42)

# Crear un imputador KNN
imputer = KNNImputer(n_neighbors=1)

# Imputar los valores faltantes en el conjunto de entrenamiento y prueba
X_train_imputed = imputer.fit_transform(X_train1)
X_test_imputed = imputer.transform(X_test1)

# Ajustar el modelo de regresiÃ³n logÃ­stica en los datos imputados
model = LogisticRegression()
model.fit(X_train_imputed, y_train1)

# Evaluar el rendimiento del modelo en los datos de prueba
y_pred1 = model.predict(X_test_imputed)
accuracy = accuracy_score(y_test1, y_pred1)
print(f"Accuracy: {accuracy}")
```

Vemos que rellenando los valores missing mediante el metodo knnimputer perdemos accuracy en el modelo de logistic regression por lo tanto no quedamos con el metodo de rellenar valores missing con un -99 (valor muy distinto a los demas)
